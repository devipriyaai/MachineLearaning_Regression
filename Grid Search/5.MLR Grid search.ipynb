```python
# importing the libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[['charges']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn. model_selection import GridSearchCV
from sklearn. linear_model import LinearRegression
param_grid={'fit_intercept':[True,False],'copy_X':[True,False],'n_jobs':[None,-1],'positive':[True,False]}

grid=GridSearchCV(LinearRegression(),param_grid,refit=True,verbose=3,n_jobs=-1)

grid.fit(x_train,y_train)
    
    
```

```python
re=grid.cv_results_
grid_predictions=grid.predict(x_test)

from sklearn.metrics import r2_score
r_score=r2_score(y_test,grid_predictions)
print("The_R score value best parameter{}:".format(grid.best_params_),r_score)
```
